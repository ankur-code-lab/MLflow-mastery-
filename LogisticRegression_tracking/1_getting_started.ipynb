```python
from sklearn import linear_model
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from mlflow.models import infer_signature
from mlflow import set_tracking_uri, \
                    set_experiment, \
                    start_run, \
                    log_params, \
                    log_metric, \
                    set_tag, \
                    sklearn, \
                    models, \
                    pyfunc
import pandas as pd
```

Load the dataset

```python
X,y = datasets.load_iris(return_X_y=True)
```

Split the data into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

Define the model hyperparameters

```python
help(linear_model.LogisticRegression)
```

```python
params = {"C": 1.0, 
"l1_ratio":0.0, 
"dual": False, 
"tol":0.0001, 
"fit_intercept":True, 
"intercept_scaling":1, 
"class_weight":None, 
"random_state":None, 
"solver":'lbfgs', 
"max_iter":100, 
"verbose":0, 
"warm_start":False, 
"n_jobs":None}
```

Model Training

```python
model = linear_model.LogisticRegression(**params)
```

```python
model.fit(X_train,y_train)
```

Model Prediction

```python
y_pred = model.predict(X_test)
y_pred
```

Model Accuracy

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

MLflow Tracking

```python
set_tracking_uri(uri="http://127.0.0.1:5000")
```

```python
set_experiment("MLFlow Quickstart")
```

Start and MLFlow Run

```python
with start_run():
    ## log hyperparameters
    log_params(params)

    ## log the metrics
    log_metric("accuracy", accuracy)

    ## set tag to remind ourselves what this run was for
    set_tag("training info", "Basic logistic regression model for IRIS dataset")

    ## Infer the model signature
    signature = infer_signature(X_train,model.predict(X_train))

    ## log the model
    model_info=sklearn.log_model(
        sk_model=model,
        artifact_path="iris_model",
        signature=signature,
        input_example=X_train,
        registered_model_name="tracking-quickstart", # you should set this parameter only if you have validated the model performance and decided that it needs to be registered

    )

```

## Model Inferencing

### Inferencing model artifacts with MLFlow Inferencing (Method 1)

```python
import json, os
from mlflow.artifacts import download_artifacts
from mlflow import models
```

```python
model_info.model_uri 
```

Download the base model_uri artifact directory in the tmp directory

```python
model_dir = download_artifacts(artifact_uri=model_info.model_uri)
model_dir
```

Join the downloaded local directory path with the file name

```python
example_file_path = os.path.join(
    model_dir, 
    model_info.saved_input_example_info['serving_input_path']
)
```

```python
example_file_path
```

Read the JSON payload string from the file

```python
with open(example_file_path, "r") as f:
    serving_payload = f.read()
```

```python
serving_payload
```

Run validation on the payload

```python
predicted_output = models.validate_serving_input(
    model_uri=model_info.model_uri, 
    serving_input=serving_payload
)
```

```python
predicted_output
```

### Load the model back for prediction as a generic python function model (Method 2)

```python
loaded_model=pyfunc.load_model(model_info.model_uri)
predictions=loaded_model.predict(X_test)

iris_features_name=datasets.load_iris().feature_names

result=pd.DataFrame(X_test,columns=iris_features_name)
result["actual_class"]=y_test
result["predcited_class"]=predictions
result
```
